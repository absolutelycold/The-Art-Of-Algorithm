# 算法内功

## 1. 时间复杂度
时间复杂度有三种计量方法。分别为$O, \Theta, \Omega$

| 标志 | 意义 |
| :--: | :--: |
| $O$ | 算法最糟糕的情况，花的最多的时间 |
| $\Theta$ | 平均情况，算法花的平均时间 |
| $\Omega$ | 最佳情况，通常为1。所以这个计量方法在计量算法时间时没有太大意义。 |

## 2. 近似记法
在实际生活中，我们需要快速得出时间复杂度。

所以我们不需要使用数学完整推理出时间复杂度。

近似记法就已足够。

近似记法就是只取出多项式中增长最快的的那一项，放到$O$中。下图给出了常见的时间复杂度:

![TimeComplexity](./Time-Complexity.png)

下面给出一些例子:

| 多项式 | 近似记法 |
| :--: | :--: |
| $$ 5n^2+2n+3log(n)$$ | $$ O(n^2) $$ |
| $$ 5n + 2log(n) + 6  $$ | $$ O(n) $$ |
| $$ 9n^3 + 2n^2 + 90nlogn + 8n + 100 $$ | $$ O(n^3) $$ |

## 3.Master Theorem(主定理)

这个定理通常用在计算递归的时间复杂度。

递归问题就是把大问题拆分为若干个小问题。

Master Theorem 定义大概如下:

Master Theorem 是用来判断形如 $ T(n) = aT(\frac{n}{b}) + f(n) $ 的时间复杂度

| 条件 | 时间复杂度$T(n)$ |
| :--: | :--: |
| $$ f(n) < n^{log_b{a}} $$ | $$ O(n^{log_b{a}}) $$ |
| $$ f(n) > n^{log_b{a}} $$ | $$ O(f(n)) $$ |
| $$ f(n) = n^{log_b{a}} $$ | $$ O(f(n)logn) $$ |

就拿**二分法**为例子:

二分法之后，input的大小为 $\frac{n}{2}$; 子任务的个数为1(只有一边需要进行下一次二分); 子任务的时间复杂度为1(只需要进行一次比较);

所以，二分法的时间复杂度可以写成: $T(n) = 1 \times T(\frac{n}{2}) + 1$

其中, a = 1, b = 2, 所以我们有 $n^{log_b{a}} = n^{log_2 1} = n^0 $; $ f(n) = 1 = n^ 0 $

$ \therefore $ $ n^{log_b{a}} = f(n) $

$ \therefore $ $ T(n) = O(f(n)logn) = O(logn) $

$ \therefore $ 我们可以得知二分法的时间复杂度为 $ O(logn) $
